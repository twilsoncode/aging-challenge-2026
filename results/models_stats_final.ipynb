```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr, spearmanr

# 1. Read the CSV files
df_submission = pd.read_csv('multi_celltype_ensemble_submission.csv')
df_true_ages = pd.read_csv('true_test_ages.csv')

# 2. Merge the datasets on 'donor_id'
df_merged = pd.merge(df_submission, df_true_ages, on='donor_id', how='inner')

# 3. Define the predicted age columns
model_cols = [
    'age_overall', 'age_CD4_T', 'age_CD8_T', 
    'age_NK', 'age_B_cells', 'age_monocytes'
]

results = []

# 4. Calculate statistics for each model column compared to the true age
for col in model_cols:
    true_age = df_merged['age']
    pred_age = df_merged[col]
    
    # Calculate metrics
    mae = mean_absolute_error(true_age, pred_age)
    rmse = np.sqrt(mean_squared_error(true_age, pred_age))
    pearson_r, _ = pearsonr(true_age, pred_age)
    spearman_rho, _ = spearmanr(true_age, pred_age)
    r2 = r2_score(true_age, pred_age)
    
    # Append to results
    results.append({
        'Model': col,
        'MAE': round(mae, 4),
        'RMSE': round(rmse, 4),
        'Pearson r': round(pearson_r, 4),
        'Spearman ρ': round(spearman_rho, 4),
        'R²': round(r2, 4)
    })

# 5. Display the results in a DataFrame format
results_df = pd.DataFrame(results)
print(results_df.to_string(index=False))
```
